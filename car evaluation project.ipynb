```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv("C:/car_evaluation.csv")
```

```python
df.head()
```

```python
df.shape
```

```python
column_names = ['Price of the car' , 'Cost of Maintenance' , 'Number of doors' , 'Number of persons' , 'Luggage space' , 'Safety Rating' , 'Car Acceptability']
```

```python
df.columns = column_names
```

```python
df.head(50)
```

```python
df['Car Acceptability'].unique()
```

```python
df['Car Acceptability'].replace('unacc', 'Unacceptable', inplace=True)
```

```python
df['Car Acceptability'].replace('acc', 'Acceptable', inplace=True)
```

```python
df['Car Acceptability'].replace('vgood', 'Very Good', inplace=True)
```

```python
df['Car Acceptability'].replace('good', 'Good', inplace=True)
```

```python
df.head()
```

```python
df.info()
```

```python
df['Car Acceptability'].value_counts()
```

```python
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder ()
```

```python
X = df.drop(['Car Acceptability'], axis=1)

y = df['Car Acceptability']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.33, random_state = 42)
```

```python
X_train.head()
```

```python
columns_names =  ['Price of the car' , 'Cost of Maintenance' , 'Number of doors' , 'Number of persons' , 'Luggage space' , 'Safety Rating']
```

```python
X_train[columns_names] = X_train[columns_names].apply(LabelEncoder().fit_transform)
```

```python
X_train.head()
```

```python
X_test[columns_names] = X_test[columns_names].apply(LabelEncoder().fit_transform)
```

```python
X_test.head()
```

```python
from sklearn.tree import DecisionTreeClassifier
classifier = DecisionTreeClassifier(criterion = 'gini', max_depth = 3 , random_state = 0)
classifier.fit(X_train, y_train)
```

```python
y_pred = classifier.predict(X_test)
```

```python
print(y_pred)
```

```python
from sklearn.metrics import accuracy_score

print('Model accuracy score with criterion gini index: {0:0.4f}'. format(accuracy_score(y_test, y_pred)))
```

```python
from sklearn.metrics import confusion_matrix,classification_report,plot_confusion_matrix
```

```python
plot_confusion_matrix(classifier,X_test,y_test)
```

```python
print(classification_report(y_test,y_pred))
```
